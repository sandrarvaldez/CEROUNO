## Es una playlist bailable?

#### Donde?
danza=spotify['danceability']

#### Como ?

- Poner un parametro de restriccion.
    - Si es menos de 0.5, no es bailable.
 
### Respuestas

Cuantas canciones son bailables usando una restriccion D?

```python
#Generacion de Panda (mi panda es SPOTIFY csv), Aqui estoy tomando solo una columna de mi panda:
danza=spotify['danceability']
#print(danza)

## pregunta 1

#Parametro de restriccion
D=0.5

#Total de canciones
C=danza.count()

#o usar LEN
C1=len(danza)

# Numero de CANCIONES BAILABLES metodo: sum-sum
C_B=sum(danza>D)

#proporcion de canciones bailables en la lista
P_B=C_B/C


# aqui en los PRINT puedo usar "," o "+" para separar la que quiero imprimir
#OJO se puede imprimir C o C1 (los 2 funcionan)
print('Existen '+str(C_B)+' canciones que son bailables de un total de '+str(C1))
#print('Existen ',str(C_B),' canciones que son bailables de un total de ',str(C1))

#OJO el ROUND es para redondear, si pongo 4 decimales me pone 2 digitos del % y otros 2 decimales.
print('El '+str(round(P_B,4)*100)+'% de las canciones son bailables segun un parametro D='+str(D))
```

```python
# Grafica de Bailable

#Matplotli.pyplot mejor conocido como plt ... lo que esta dentro de los parentesis son los argumentos, 30 barritas
plt.hist(danza,30,facecolor="green")

# se pone ua en bailabilidad porque no tenemos una unidad que medir no son cm, metros, etc, es una UNIDAD ARBITRARIA
plt.xlabel('Bailabilidad [ua]')
print()
plt.ylabel('Numero de canciones ')
plt.title('Numero de canciones respecto a la bailabilidad ')
#para que nadamas me muestre la grafica y no me muestre los datos se pone este print, es importante que el show hasta el final.
plt.show()

#mi grafica muestra por ejemplo en este rango tengo tanto numero de canciones
```

```python
danza.plot(kind='hist',bins=30)
plt.xlabel('Bailabilidad [ua]')
print()
plt.ylabel('Numero de canciones ')
plt.title('Numero de canciones respecto a la bailabilidad ')
plt.show()
```

# Relacion entre bailable y energia

```python
#Generacion de Panda (mi panda es SPOTIFY csv), Aqui estoy tomando solo una columna de mi panda:
#energia=spotify['energy']

spotify.plot.scatter('danceability','energy')
plt.show()
```

```python
## TENDENCIA DE DURACION PREGUNTA 4

duracion=spotify['duration_ms']
duracion=duracion*.001/60
duracion.plot(kind='hist',bins=30)

#Esto lo hizo lorena
promedio=sum(duracion)/len(duracion)
print("El promedio de canciones mas escuchadas es {} minutos en {} canciones analizadas".format(round(promedio,4),len(duracion)))

plt.xlabel('duracion [minutos]')
print()
plt.ylabel('Numero de canciones ')
plt.title('Duracion de las canciones ')
plt.show()
```

```python
## TENDENCIA DE SPEECHINESS PREGUNTA 5

vocalizacion=spotify['speechiness']
vocalizacion.plot(kind='hist',bins=30)

promedio=sum(vocalizacion)/len(vocalizacion)
print("El promedio de canciones mas vocalizadas es {} en {} canciones analizadas".format(round(promedio,4),len(duracion)))

plt.xlabel('Palabras')
print()
plt.ylabel('Numero de canciones ')
plt.title('Vocalizacion ')
plt.show()
```

```python
## ES ESTA UNA LISTA DE CONCIERTOS?  PREGUNTA 6

en_vivo=spotify['liveness']
en_vivo.plot(kind='hist',bins=30,facecolor='purple')

promedio=sum(en_vivo)/len(en_vivo)
print("El promedio de canciones en vivo es {} en {} canciones analizadas".format(round(promedio,4),len(en_vivo)))

plt.xlabel('Tocadas en vivo')
print()
plt.ylabel('Numero de canciones ')
plt.title('Canciones tocadas en vivo ')
plt.show()
```

```python
## TIENE TEMPO LENTO O RAPIDO?  PREGUNTA 7

ritmo=spotify['tempo']
ritmo.plot(kind='hist',bins=30,facecolor='green')

promedio=sum(ritmo)/len(ritmo)
print("El promedio de canciones en vivo es {} en {} canciones analizadas".format(round(promedio,4),len(ritmo)))

plt.xlabel('Ritmo de las canciones')
print()
plt.ylabel('Numero de canciones ')
plt.title('Ritmo de las Canciones')
plt.show()
```
